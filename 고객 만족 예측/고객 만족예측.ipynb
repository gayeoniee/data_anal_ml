```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.ensemble import GradientBoostingClassifier
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings(action='ignore')
```

# 데이터 확인

```python
cust_df = pd.read_csv('Customertrain.csv')
cust_df.head()
```

```python
# 컬럼이 371 -> 고차원 데이터
cust_df.info()
```

# 데이터 전처리

```python
cust_df['var3'].value_counts()
```

```python
# -999999 라는 이상치를 최빈값 2로 대체하여 처리
cust_df['var3'].replace(-999999, 2, inplace=True)
```

```python
# null 값이 있는 행 확인
na = cust_df.isnull().mean().sort_values(ascending=False)
na[na > 0]
```

```python
cust_df.dtypes.value_counts()
```

```python
# 상관관계를 보고
# 95% 가 넘는 관계를 가진 컬럼 삭제
corr = cust_df.corr().abs()
upper = corr.where(np.triu(np.ones(corr.shape), k=1).astype(bool))
to_drop = [column for column in upper.columns if any(upper[column] > 0.95)]
cust_df.drop(columns=to_drop, inplace=True)
```

```python
cust_df.columns
```

# 데이터 나누기

```python
X = cust_df.drop(columns=['ID', 'TARGET'])
y = cust_df['TARGET']

print(y.value_counts())
```

0과 1의 비율이 매우 차이남을 알 수 있음

```python
# 
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)

train_cnt = y_train.count()
test_cnt = y_test.count()
print('학습 세트 Shape:{0}, 테스트 세트 Shape:{1}'.format(X_train.shape , X_test.shape))

print('train')
print(y_train.value_counts()/train_cnt)
print('\n test')
print(y_test.value_counts()/test_cnt)
```

학습 데이터와 테스트 데이터의 비율이 비슷하게 나뉘었는지 확인

# 모델 성능 비교

```python
from sklearn.metrics import roc_auc_score
results = []

models = {
    "Logistic Regression": LogisticRegression(max_iter=1000),
    "Random Forest": RandomForestClassifier(n_estimators=100, random_state=42),
    "XGBoost": XGBClassifier(eval_metric='auc', random_state=42),
    "Gradient Boosting": GradientBoostingClassifier(random_state=42)
}

for name, model in models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    
    roc = roc_auc_score(y_test, y_pred, average='macro')
    print(f"\n[{name}]")
    print(classification_report(y_test, y_pred))
    
    results.append({'Model': name, 'AUC': round(roc, 4)})

# 결과 정리
results_df = pd.DataFrame(results).sort_values(by='AUC', ascending=False)
```

1에 대한 정밀도와 재현율이 매우 떨어지는 것을 확인 할 수 있음   

```python
from sklearn.metrics import roc_auc_score

for name, model in models.items():
    model.fit(X_train, y_train)
    
    # 확률 예측
    y_proba = model.predict_proba(X_test)[:, 1]
    
    auc = roc_auc_score(y_test, y_proba)
    print(f"[{name}] ROC AUC: {auc:.4f}")
```

캐글 대회의 평가 지표인 ROC AUC를 확인   
-> xgboost와 gradient boosting 채택

# 모델 선택

```python
xgb_model = XGBClassifier(eval_metric='auc', random_state=42)

xgb_model.fit(
    X_train, y_train,
    eval_set=[(X_test, y_test)],
    early_stopping_rounds=40,
    verbose=False
)
```

```python
y_pred_proba = xgb_model.predict_proba(X_test)[:, 1]

auc = roc_auc_score(y_test, y_pred_proba)
print(f"ROC AUC: {auc:.4f}")

y_pred = xgb_model.predict(X_test)
print(classification_report(y_test, y_pred))
```

```python
gb_model = GradientBoostingClassifier(random_state=42)

gb_model.fit(X_train, y_train)

y_pred_proba = gb_model.predict_proba(X_test)[:, 1]

auc = roc_auc_score(y_test, y_pred_proba)
print(f"ROC AUC: {auc:.4f}")

y_pred = gb_model.predict(X_test)
print(classification_report(y_test, y_pred))
```

위 두 모델의 제출 결과   
- xgb => Private Score : 0.81869 , Public Score : 0.83052

- gb => Private Score : 0.82086 , Public Score : 0.83311


# 모델 튜닝

```python
from sklearn.model_selection import GridSearchCV

params = {
    'n_estimators': [100, 200, 300],
    'learning_rate': [0.05, 0.1, 0.2],
    'max_depth': [3, 4, 5],
    'subsample': [0.8, 1.0],
    'min_samples_split': [2, 5],
}

gb_model = GradientBoostingClassifier(random_state=42)

grid = GridSearchCV(
    gb_model,
    params,
    scoring='roc_auc',
    cv=5,
    verbose=1,
    n_jobs=-1
)

grid.fit(X_train, y_train)

print("최적 하이퍼파라미터:", grid.best_params_)
print("최고 AUC 점수:", grid.best_score_)
```

```python
xgb_model = XGBClassifier(eval_metric='auc', use_label_encoder=False, random_state=42)

params = {
    'n_estimators': [100, 200],
    'learning_rate': [0.05, 0.1],
    'max_depth': [3, 4, 5],
    'subsample': [0.8, 1.0],
    'colsample_bytree': [0.8, 1.0],
    'min_child_weight': [1, 3]
}

grid_xgb = GridSearchCV(
    xgb_model,
    param_grid=params,
    scoring='roc_auc',
    cv=5,
    verbose=1,
    n_jobs=-1
)

grid_xgb.fit(X_train, y_train)

print("최적 하이퍼파라미터:", grid_xgb.best_params_)
print("최고 AUC 점수:", grid_xgb.best_score_)
```
